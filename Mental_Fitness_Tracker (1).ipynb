```
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in data.columns:
  if data[i].dtype=='object':
    data[i]=l.fit_transform(data[i])
```

```
data.shape
```

```
x=data.drop('mental_fitness',axis=1)
y=data['mental_fitness']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=.20, random_state=2)
```

```
print('x_train:',x_train.shape)
print('x_test:',x_test.shape)
print('\ny_train:',y_train.shape)
print('y_test:',y_test.shape)
```

```
# Building model for training set using Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
lr = LinearRegression()
lr.fit(x_train,y_train)

y_pred=lr.predict(x_train)

mse=mean_squared_error(y_train,y_pred)

rmse = (np.sqrt(mean_squared_error(y_train,y_pred)))

r2=r2_score(y_train,y_pred)

print("the Linear Regression model performance for training set")
print('--------------')
print('The MSE is {}'.format(mse))
print('The RMSE is {}'.format(rmse))
print('The R2 is {}'.format(r2))
```

```
# Building model for training set using Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train, y_train)
y_pred=rf.predict(x_train)

mse=mean_squared_error(y_train,y_pred)

rmse = (np.sqrt(mean_squared_error(y_train,y_pred)))

r2=r2_score(y_train,y_pred)

print("the Random FOrest Regressor model performance for training set")
print('--------------')
print('The MSE is {}'.format(mse))
print('The RMSE is {}'.format(rmse))
print('The R2 is {}'.format(r2))
```

```
# Model Evaluation with test data using Linear Regression

y_pred_test=lr.predict(x_test)

mse=mean_squared_error(y_test, y_pred_test)

rmse = (np.sqrt(mean_squared_error(y_test,y_pred_test)))

r2=r2_score(y_test,y_pred_test)

print("The Linear Regression model performance for test set:")
print('--------------')
print('The MSE is {}'.format(mse))
print('The RMSE is {}'.format(rmse))
print('The R2 is {}'.format(r2))


# Model Evaluation with test data using Random Forest Regressor

y_pred_test=rf.predict(x_test)

mse=mean_squared_error(y_test, y_pred_test)

rmse = (np.sqrt(mean_squared_error(y_test,y_pred_test)))

r2=r2_score(y_test,y_pred_test)

print("\nThe Random Forest model performance for test set:")
print('--------------')
print('The MSE is {}'.format(mse))
print('The RMSE is {}'.format(rmse))
print('The R2 is {}'.format(r2))
```

```
# Building model using Desicion Tree Regressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Example dataset (features and continuous values)

x=data.drop('mental_fitness',axis=1)
y=data['mental_fitness']
a_train, a_test, b_train, b_test=train_test_split(x, y, test_size=.20, random_state=2)
# Train the regressor
regressor = DecisionTreeRegressor()
regressor.fit(a_train, b_train)

# Make predictions on the test set
b_pred = regressor.predict(a_train)

# Calculate the mean squared error (MSE)
mse = mean_squared_error(b_train, b_pred)

rmse = (np.sqrt(mean_squared_error(b_train,b_pred)))

r2=r2_score(b_train,b_pred)
print('The MSE is {}'.format(mse))
print('The RMSE is {}'.format(rmse))
print('The R2 is {}'.format(r2))
```

```
# Model Evaluation with test data using Decision Tree Regressor

b_pred_test=regressor.predict(a_test)

mse=mean_squared_error(b_test, b_pred_test)

rmse = (np.sqrt(mean_squared_error(b_test,b_pred_test)))

r2=r2_score(b_test,b_pred_test)

print("\nThe Decision Tree Regressor model performance for test set:")
print('--------------')
print('The MSE is {}'.format(mse))
print('The RMSE is {}'.format(rmse))
print('The R2 is {}'.format(r2))
```
